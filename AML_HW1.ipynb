1. What is the difference between R-square and adjusted R square and why is it desirable to use the adjusted value? (4 pts)

2. Overfitting usually happens in complex models. Linear Regression is a fairly simple model. Could overfitting happen in Linear Regression? If so, please explain the scenario in which it could happen and how we can tackle it. (6 pts)


Ans:
1. R-squared (R) and adjusted R-squared are both metrics used to evaluate the goodness of fit of a regression model. However, they have distinct purposes and interpretations.

R-squared (R) measures the proportion of the variance in the dependent variable that is explained by the independent variables in the model. It ranges from 0 to 1, where a higher value indicates a better fit. However, R-squared tends to increase with the addition of more variables, even if they do not contribute significantly to the model's explanatory power.

Adjusted R-squared, on the other hand, takes into account the number of independent variables in the model. It penalizes the inclusion of unnecessary variables that might not contribute meaningfully to the model's prediction performance. Adjusted R-squared is particularly useful when comparing models with different numbers of variables. It provides a more realistic assessment of the model's explanatory power and helps prevent overfitting by accounting for the potential inclusion of irrelevant variables.

In summary, adjusted R-squared is preferred over R-squared when dealing with multiple independent variables, as it offers a balanced evaluation of model complexity and explanatory power.





2.  While Linear Regression is a relatively simple model, overfitting can still occur under certain circumstances. Overfitting happens when a model captures noise and random fluctuations in the training data, leading to poor generalization on new, unseen data.

In Linear Regression, overfitting can occur if the model becomes too complex relative to the amount of available training data. This can happen in scenarios where:
- The number of predictor variables is significantly larger than the number of data points.
- Polynomial features of high degrees are included, causing the model to fit the noise instead of the underlying trend.

To tackle overfitting in Linear Regression:
1. **Feature Selection:** Choose relevant and meaningful predictor variables. Avoid including variables that do not contribute substantively to the model's predictive power.
2. **Regularization Techniques:** Apply regularization methods such as Ridge Regression or Lasso Regression. These techniques add penalty terms to the loss function, discouraging overly large coefficients and reducing the model's complexity.
3. **Cross-Validation:** Employ techniques like k-fold cross-validation to assess the model's performance on different subsets of data. This helps to identify if the model is overfitting by checking its consistency across folds.
4. **Early Stopping:** If the model's performance starts degrading on a validation set during training, stop training to prevent further overfitting.

In conclusion, while Linear Regression is less prone to overfitting compared to complex models, it can still occur if the model becomes too intricate relative to the data available. Careful model selection, regularization, and cross-validation are essential for mitigating this issue.

Use `sklearn.linear_model.Lasso` and `sklearn.linear_model.Ridge` classes to do a 5-fold cross validation using sklearn's `KFold`. For the sweep of the regularization parameter, we will look at a grid of values ranging from  $\alpha=10^{-6}$  to  $\alpha=10^{6}$.In Python, you can consider this range of values as follows: `alpha = 10**numpy.linspace(-6, 6, 100)` so that you can generate 100 uniform values between -6 to 6 as power series.

Fit the 2 regression models (Lasso and Ridge) with scaled data and report the best chosen $\alpha$ based on cross validation as well as the corresponding scoring metric. The cross validation should happen on your training data using MSE as the scoring metric.

```python
# Define number of folds
##  START CODE  ## (1 line of code)
n_folds = 5
##  END CODE  ##
```

```python
# Create KFold from sklearn
##  START CODE  ## (1 line of code)
k_fold = KFold(n_splits=n_folds)
##  END CODE    ##
```

```python
#Define the alphas as defined in the question
##  START CODE  ## (1 line of code)
alphas = 10**np.linspace(-6, 6, 100)
##  END CODE    ##

lasso_avg_mse = {}
```

```python
#For each value of alpha and each fold compute the mean square error
for alpha in alphas:

  #Instantiate a lasso model with the current alpha
  ##  START CODE  ## (1 line of code)
  lasso = linear_model.Lasso(alpha=alpha)
  ##  END CODE    ##

  avg_mse = 0

  for k, (train, test) in enumerate(k_fold.split(Xscaled_train, y_train)):

    #Fit the scaled training data to the lasso model
    ## START CODE ## (1 line of code)
    lasso.fit(Xscaled_train[train], y_train[train])
    ## END CODE ##

    #Calculate the average mean sqaured error
    ##  START CODE  ## (1 line of code)
    avg_mse = avg_mse + mean_squared_error(y_train[test], lasso.predict(Xscaled_train[test]))
    ##  END CODE    ##

  # Take the average mean squared error as metric
  lasso_avg_mse[alpha] = avg_mse / n_folds
```

```python
# Find the best value for alpha with minimum mean squared error
##  START CODE  ## (1 line of code)
best_alpha_lasso = min(lasso_avg_mse, key=lasso_avg_mse.get)
##  END CODE    ##

print("Best lasso alpha: {}".format(best_alpha_lasso))
```

```python
ridge_avg_mse = {}
#For each value of alpha and each fold compute the mean square error
for alpha in alphas:

  #Instantiate a ridge model with the current alpha
  ##  START CODE  ## (1 line of code)
  ridge = linear_model.Ridge(alpha=alpha)
  ##  END CODE    ##

  avg_mse = 0

  for k, (train, test) in enumerate(k_fold.split(Xscaled_train, y_train)):

    #Fit the scaled training data to the ridge model
    ## START CODE ## (1 line of code)
    ridge.fit(Xscaled_train[train], y_train[train])
    ## END CODE ##

    #Calculate the average mean sqaured error
    ##  START CODE  ## (1 line of code)
    avg_mse = avg_mse + mean_squared_error(y_train[test], ridge.predict(Xscaled_train[test]))
    ##  END CODE    ##

  # Take the average mean squared error as metric
  ridge_avg_mse[alpha] = avg_mse / n_folds
```

```python
# Find the best value for alpha with minimum mean squared error
##  START CODE  ## (1 line of code)
best_alpha_ridge = min(ridge_avg_mse, key=ridge_avg_mse.get)
##  END CODE    ##

print("Best Ridge alpha: {}".format(best_alpha_ridge))
```

When the value of the regularization parameter (alpha) changes in Ridge and Lasso regression, we can qualitatively observe the following effects on the learned coefficients:

**1. Lasso Regression:**

- **Increasing Alpha**: As alpha increases, Lasso tends to push the coefficients of less important features towards zero more aggressively. This results in feature selection, where some coefficients become exactly zero, effectively removing those features from the model. It promotes sparsity in the model, keeping only the most relevant features.

- **Decreasing Alpha**: As alpha decreases, Lasso becomes less restrictive, allowing more features to have non-zero coefficients. The model becomes more complex and may overfit the training data if alpha is too small.

**2. Ridge Regression:**

- **Increasing Alpha**: As alpha increases, Ridge shrinks the coefficients of all features towards zero. However, it does not force any coefficients to become exactly zero. Ridge tends to maintain all features' relevance but penalizes large coefficients, preventing overfitting.

- **Decreasing Alpha**: As alpha decreases, Ridge becomes less restrictive, allowing the coefficients to grow larger. The model becomes more flexible but may be prone to overfitting if alpha is too small.

In summary, the regularization parameter (alpha) controls the balance between model complexity and the fitting of the data. Higher alpha values lead to more regularization and simpler models with fewer features (Lasso) or smaller coefficients (Ridge), while lower alpha values allow the model to be more flexible and potentially overfit. The choice of alpha should be based on the trade-off between bias and variance, and it often requires tuning through techniques like cross-validation to find the optimal value for your specific dataset.

Observe the plot, then explain the pros and cons of ridge, lasso and Elastic Net models.

Observing the coefficient plots for Ridge, Lasso, and Elastic Net models can provide insights into the pros and cons of each:

**Ridge Regression:**
- **Pros:**
  - Ridge regression is effective when there is multicollinearity among the independent variables. It tends to distribute the coefficients among correlated variables, which can help stabilize the model.
  - Ridge regression does not force coefficients to become exactly zero, allowing it to retain all features. This can be advantageous when you believe all features are relevant to the prediction.

- **Cons:**
  - Ridge may not perform feature selection; it tends to keep all features but shrink their coefficients. If some features are truly irrelevant, Ridge may not provide a clear indication of that.
  - It may not perform as well as Lasso when there are many irrelevant features and you want a more parsimonious model.

**Lasso Regression:**
- **Pros:**
  - Lasso performs automatic feature selection by setting some coefficients to exactly zero. This makes the model interpretable and helps identify the most important features.
  - It is effective when dealing with a large number of features, especially when many of them are irrelevant, as it can create a sparse model.

- **Cons:**
  - Lasso can be sensitive to the scale of features. Standardization or scaling of features is often required.
  - When there is multicollinearity among features, Lasso may arbitrarily select one of the correlated features and set the others to zero, which can be unstable.

**Elastic Net Regression:**
- **Pros:**
  - Elastic Net combines the advantages of Ridge and Lasso. It can handle multicollinearity like Ridge and perform feature selection like Lasso.
  - It provides a balance between the L1 and L2 regularization, which allows it to address the limitations of both Ridge and Lasso.

- **Cons:**
  - Elastic Net has an additional hyperparameter, alpha, that needs to be tuned, which adds complexity to the model selection process.
  - It may still struggle with feature selection in cases of strong multicollinearity, although it is less likely to exclude correlated variables than Lasso.

In summary, the choice between Ridge, Lasso, and Elastic Net depends on the specific characteristics of your dataset and the goals of your modeling:

- Use **Ridge** when you want to handle multicollinearity and believe that most features are relevant but want to shrink their coefficients to avoid overfitting.

- Use **Lasso** when you want automatic feature selection and believe that many features are irrelevant, leading to a sparse model.

- Use **Elastic Net** when you want a balance between Ridge and Lasso, which can be useful in situations where both multicollinearity and feature selection are concerns.

It's often beneficial to experiment with different models and alpha values using techniques like cross-validation to determine which regularization approach works best for your specific problem.

## Part-7: *(3 pts)*

Why is it advisable to scale the independent variables when applying ridge or  lasso regression?

Scaling the independent variables is advisable when applying Ridge or Lasso regression because-

1. Ridge and Lasso regression penalize the magnitude of coefficients to prevent overfitting. If the variables have different scales, the algorithm may assign higher penalties to variables with larger scales. Scaling ensures that all variables are on a similar scale, preventing this bias in regularization.

2. When comparing the magnitudes of coefficients in the resulting model, scaling makes it easier to interpret the importance of each feature. Without scaling, it may be challenging to determine which features are more or less important based solely on the magnitude of their coefficients

3. It allows for better comparability of model performance metrics. If features are on different scales, comparing model performance across different models or datasets can be challenging
